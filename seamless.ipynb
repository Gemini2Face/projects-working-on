```python
import whisper
from transformers import MarianMTModel, MarianTokenizer

whisper_model = whisper.load_model("large")

model_name = "Helsinki-NLP/opus-mt-en-de"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

def transcribe_and_translate_audio_whisper(audio_path, target_language="de"):
    result = whisper_model.transcribe(audio_path, task="translate", language=target_language)
    return result["text"]

def transcribe_audio_whisper(audio_path):
    result = whisper_model.transcribe(audio_path)
    return result["text"]

def translate_text_marian_mt(text, target_language="de"):
    # Prepare the text for translation
    translated = model.generate(**tokenizer(text, return_tensors="pt", padding=True))
    translation = [tokenizer.decode(t, skip_special_tokens=True) for t in translated]
    return translation[0]

def process_audio_file(audio_file_path):
    transcription_whisper = transcribe_audio_whisper(audio_file_path)
    print("Transcription (Whisper):")
    print(transcription_whisper)
    
    translation_marian_mt = translate_text_marian_mt(transcription_whisper, "de")
    print("\nTranslation (MarianMT):")
    print(translation_marian_mt)
    
    translation_whisper = transcribe_and_translate_audio_whisper(audio_file_path, "de")
    print("\nDirect Translation (Whisper):")
    print(translation_whisper)
    
    return transcription_whisper, translation_marian_mt, translation_whisper

if __name__ == "__main__":
    audio_file_path = r'C:\Users\abhis\OneDrive\Desktop\realign voice cloning\Voicy_Answer your question for silence.wav'
    transcription, translation_marian_mt, translation_whisper = process_audio_file(audio_file_path)
```

```python
from transformers import MarianMTModel, MarianTokenizer
from datasets import load_metric
import sacrebleu
from rouge_score import rouge_scorer

model_name = "Helsinki-NLP/opus-mt-en-de"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

bleu_metric = load_metric("sacrebleu", trust_remote_code=True)
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

def translate_text_marian_mt(texts, target_language="de"):
    translated_texts = []
    for text in texts:
        translated = model.generate(**tokenizer(text, return_tensors="pt", padding=True))
        translation = [tokenizer.decode(t, skip_special_tokens=True) for t in translated]
        translated_texts.append(translation[0])
    return translated_texts

def evaluate_translations(predictions, references):
    bleu = sacrebleu.corpus_bleu(predictions, [references])
    print(f"BLEU score: {bleu.score}")

    rouge1 = rouge2 = rougeL = 0
    for pred, ref in zip(predictions, references):
        scores = scorer.score(ref, pred)
        rouge1 += scores['rouge1'].fmeasure
        rouge2 += scores['rouge2'].fmeasure
        rougeL += scores['rougeL'].fmeasure

    n = len(predictions)
    print(f"ROUGE-1 score: {rouge1 / n}")
    print(f"ROUGE-2 score: {rouge2 / n}")
    print(f"ROUGE-L score: {rougeL / n}")

if __name__ == "__main__":
    source_texts = [
        "This is a test sentence.",
        "How are you doing today?",
        "Machine translation is a complex task."
    ]
    reference_translations = [
        "Dies ist ein Testsatz.",
        "Wie geht es Ihnen heute?",
        "Maschinelle Übersetzung ist eine komplexe Aufgabe."
    ]

    translations = translate_text_marian_mt(source_texts, "de")

    print("Translations:")
    for i, translation in enumerate(translations):
        print(f"Source: {source_texts[i]}")
        print(f"Translation: {translation}")
        print(f"Reference: {reference_translations[i]}")
        print()

    evaluate_translations(translations, reference_translations)
```
